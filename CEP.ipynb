```python
import pandas as pd
import requests
import pprint
```

```python
cep = 11663008
# cep = cep.replace("-", "").replace(".", "").replace(" ", "")  *** Editar todos os espaços, hifens e traços indevidos deste campo CEP
# if len(cep) == 8                                                           *** Verificar se o campo CEP possui no máximo 8 números , caso contrário não faça a requisição

link = f"https://viacep.com.br/ws/{cep}/json/"
conteudo = requests.get(link)
cepjson = conteudo.json()


# if condição : , elif condição : , else condição :


if conteudo.status_code == 400 :
  print("Resposta *cep ñ existe*")
  print(link, " = ", conteudo.status_code)

else :

  if conteudo.status_code == 200 :
    print("Resposta Ok = ")
    print(link, " = ", conteudo.status_code)
    #print(conteudo.json())
    for dados in conteudo.json():
      print(dados, " : ", cepjson[f'{dados}'])


  else :
    print("Resposta *rejeitada*")
    print(link, " = ", conteudo.status_code)


```
